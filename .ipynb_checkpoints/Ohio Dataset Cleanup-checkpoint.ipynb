ohio_completion has actual opening year data. 

diffPOY_NAOY= Planned Opening Year-Near Actual Open Year 

diffPOY_RY=Recent Year-Planned Opening Year

diffPOY-AOY= Planned OpeningYear-Actual OpeningYear

diffAOY_NAOY=ActualOpening year- Near Actual Open Year

## Explanation

If Planned Opening Year (2005) is greater than Actual Opening Year (2004), that means the project is planned for 2005 but opened in 2004. Forecast is available for 2005 but the quesiton is which actual count to take.

If the Near Actual Open Year (2003) is less than Actual Opening Year (2004), that means Near Actual Open Year Count is before the project has been opened and therefore cannot be compared against. In this case the actual year should be Recent Year. 
Scale Forecast to Recent Year.

If Near Actual Open Year (2005) is greater than Actual Opening Year (2004), that means the count can be used. But compare diffAOY_NAOY with diffRY_AOY. The smaller value should be the year.


```python
#working on the ohio_completion
#need to see if NearActualOpeningYear and RecentYear are after the project completion year or not.

#ohio_completion.describe()

ohio_cleaned=ohio_completion[ohio_completion['diffPOY_AOY']>=0]
#in this dataset we have only the projects that are built early or on time
ohio_cleaned
```

```python
def Year(x):
    if x.diffAOY_NAOY>0: 
        year=x.RecentYr
    else:
        if x.diffAOY_RY>x.diffAOY_NAOY:
            year=x.NearActualOpenYear
        else:
            year=x.RecentYr
    return year

def TrafficForecast(x):
    m=(x.DYADTForecast-x.OYADTForecast)/(x.DesignYear-x.OpeningYear)
    adjustedforecast=m*(x.ForecastYear-x.OpeningYear)+x.OYADTForecast
    return adjustedforecast
    
#YourDataFrame['NewColumn'] = YourDataFrame.apply(Year, axis = 1)
```

```python
ohio_completionx=ohio_completionx.assign(ForecastYear=ohio_completionx.apply(Year, axis=1))
ohio_completionx
```

```python
ohio_completionx=ohio_completionx.assign(TrafficForecast=ohio_completionx.apply(TrafficForecast, axis=1))
TrafficForecast
```

# Adjust Type 1
## Scale up the forecast to recent Year.
### use y=mx+c
$m=\frac{OYADTForecast-DYADTForecast}{OpeningYear-DesignYear}$

$adjusted count=m*(RecentYr-OpeningYear)+OYADTForecast$

```python
ohio_scaletype1=ohio_scaletype1.assign(m=(ohio_scaletype1['DYADTForecast']-ohio_scaletype1['OYADTForecast'])/
                                            (ohio_scaletype1['DesignYear']-ohio_scaletype1['OpeningYear']))
ohio_scaletype1[ohio_scaletype1['m']==0].head()
```

```python
ohio_scaletype1=ohio_scaletype1.assign(AdjustedForecast=ohio_scaletype1['m']*(ohio_scaletype1['RecentYr']-ohio_scaletype1['OpeningYear'])
                                      +ohio_scaletype1['OYADTForecast'])
ohio_scaletype1.head()
```
