```python
import pandas as pd #for working with dataframe
import numpy as np
from scipy import stats, integrate #will help in stats operation in histograms
import matplotlib.pyplot as plt #to plot the PE
import seaborn as sns #nice looking plots
import statsmodels.formula.api as smf
%matplotlib inline
sns.set(color_codes=True)
```

```python
flatdata_adjusted=pd.read_csv('flatdata_adjusted2.csv')
flatdata_adjusted
```

```python
flatdata_adjusted=flatdata_adjusted[flatdata_adjusted['AbsPE']<150]
```

```python
flatdata_adjusted=flatdata_adjusted.assign(DiffYear=flatdata_adjusted['ForecastYear']-flatdata_adjusted['YearForecastProduced'])
```

```python
def UnifiedImprovement(x): #to adjust the forecast
    if x.ImprovementType<7:
        unifiedtype=1
    elif x.ImprovementType<12:
        unifiedtype=2
    else:
        unifiedtype=3
    return unifiedtype
```

```python
flatdata_adjusted=flatdata_adjusted.assign(UnifiedImprovementType=flatdata_adjusted.apply(UnifiedImprovement, axis=1))
```

```python
list(flatdata_adjusted)
```

```python
pd.options.display.max_columns=50
```

```python
mod = smf.ols(formula='Count \
                       ~ AdjustedForecast', data=flatdata_adjusted)
res = mod.fit()
print(res.summary())
```

```python
mod = smf.ols(formula='Count \
                       ~ AdjustedForecast \
                       - 1', data=flatdata_adjusted)
res = mod.fit()
print(res.summary())
```

```python
mod = smf.ols(formula='Count \
                       ~ AdjustedForecast \
                       +DiffYear \
                       +FunctionalClass \
                       +ImprovementType \
                       +ForecastMethodology', data=flatdata_adjusted)
res = mod.fit()
print(res.summary())
```

```python
mod = smf.ols(formula='Count \
                       ~ AdjustedForecast \
                       +DiffYear \
                       +C(FunctionalClass) \
                       +C(UnifiedImprovementType) \
                       +C(ForecastMethodology)', data=flatdata_adjusted)
res = mod.fit()
print(res.summary())
```

```python
mod = smf.ols(formula='Count \
                       ~ AdjustedForecast \
                       +C(ForecastMethodology)', data=flatdata_adjusted)
res = mod.fit()
print(res.summary())
```

```python
flatdata_adjusted['AdjForecast_FM2'] = np.where(flatdata_adjusted['ForecastMethodology']==2, flatdata_adjusted['AdjustedForecast'], 0)
flatdata_adjusted['AdjForecast_FM4'] = np.where(flatdata_adjusted['ForecastMethodology']==4, flatdata_adjusted['AdjustedForecast'], 0)
flatdata_adjusted['AdjForecast_FM5'] = np.where(flatdata_adjusted['ForecastMethodology']==5, flatdata_adjusted['AdjustedForecast'], 0)
flatdata_adjusted['AdjForecast_FM6'] = np.where(flatdata_adjusted['ForecastMethodology']==6, flatdata_adjusted['AdjustedForecast'], 0)

mod = smf.ols(formula='Count \
                       ~ AdjustedForecast \
                       + AdjForecast_FM2 \
                       + AdjForecast_FM4 \
                       + AdjForecast_FM5 \
                       + AdjForecast_FM6',
                       data=flatdata_adjusted)
res = mod.fit()
print(res.summary())
```

```python
flatdata_adjusted['Recession'] = np.where((flatdata_adjusted['Year_ofObservation']>=2009) & (flatdata_adjusted['Year_ofObservation']<=2014), 1, 0)
```

```python
flatdata_adjusted['AdjForecast_FM2'] = np.where(flatdata_adjusted['ForecastMethodology']==2, flatdata_adjusted['AdjustedForecast'], 0)
flatdata_adjusted['AdjForecast_FM4'] = np.where(flatdata_adjusted['ForecastMethodology']==4, flatdata_adjusted['AdjustedForecast'], 0)
flatdata_adjusted['AdjForecast_FM5'] = np.where(flatdata_adjusted['ForecastMethodology']==5, flatdata_adjusted['AdjustedForecast'], 0)
flatdata_adjusted['AdjForecast_FM6'] = np.where(flatdata_adjusted['ForecastMethodology']==6, flatdata_adjusted['AdjustedForecast'], 0)

flatdata_adjusted['AdjForecast_DiffYear'] = flatdata_adjusted['AdjustedForecast'] * flatdata_adjusted['DiffYear']

mod = smf.ols(formula='Count \
                       ~ AdjustedForecast \
                       + AdjForecast_FM4 \
                       + AdjForecast_FM5 \
                       + AdjForecast_FM6 \
                       + AdjForecast_DiffYear \
                       + Recession',
                       data=flatdata_adjusted)
res = mod.fit()
print(res.summary())
```

```python
flatdata_adjusted['AdjForecast_FM2'] = np.where(flatdata_adjusted['ForecastMethodology']==2, flatdata_adjusted['AdjustedForecast'], 0)
flatdata_adjusted['AdjForecast_FM4'] = np.where(flatdata_adjusted['ForecastMethodology']==4, flatdata_adjusted['AdjustedForecast'], 0)
flatdata_adjusted['AdjForecast_FM5'] = np.where(flatdata_adjusted['ForecastMethodology']==5, flatdata_adjusted['AdjustedForecast'], 0)
flatdata_adjusted['AdjForecast_FM6'] = np.where(flatdata_adjusted['ForecastMethodology']==6, flatdata_adjusted['AdjustedForecast'], 0)

flatdata_adjusted['AdjForecast_DiffYear'] = flatdata_adjusted['AdjustedForecast'] * flatdata_adjusted['DiffYear']

flatdata_adjusted['AdjForecast_Recession'] = flatdata_adjusted['AdjustedForecast'] * flatdata_adjusted['Recession']


mod = smf.ols(formula='Count \
                       ~ AdjustedForecast \
                       + AdjForecast_FM4 \
                       + AdjForecast_FM5 \
                       + AdjForecast_FM6 \
                       + AdjForecast_DiffYear \
                       + AdjForecast_Recession',
                       data=flatdata_adjusted)
res = mod.fit()
print(res.summary())
```
