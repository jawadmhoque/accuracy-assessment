```python
def find_PID(value): #to clean up the data
    if isinstance(value, str):
        end = value.find('--')
        if end == -1:
            return None
        else:
            start = value.find('-')+1
            return int(value[start:end])
```

```python
def Year(x): #get the year to compare
    if x.ImprovementType==1: #low risk projects
        year=x.ForecastYear+1
    else:
        year=x.ForecastYear+2 #buffer of 2 years.
    
    return year

def TrafficCount(x):
    if x.ForecastYear==2003:
        Count=x.Y2003
    elif x.ForecastYear==2004:
        Count=x.Y2004
    elif x.ForecastYear==2005:
        Count=x.Y2005
    elif x.ForecastYear==2006:
        Count=x.Y2006
    elif x.ForecastYear==2007:
        Count=x.Y2007
    elif x.ForecastYear==2008:
        Count=x.Y2008
    elif x.ForecastYear==2009:
        Count=x.Y2009
    elif x.ForecastYear==2010:
        Count=x.Y2010
    elif x.ForecastYear==2011:
        Count=x.Y2011
    elif x.ForecastYear==2012:
        Count=x.Y2012
    elif x.ForecastYear==2013:
        Count=x.Y2013
    elif x.ForecastYear==2014:
        Count=x.Y2014
    elif x.ForecastYear==2015:
        Count=x.Y2015
    else:
        Count='NA'
    return Count
```

```python
def TrafficForecast(x): #adjust the forecast
    if x.Year_ofObservation!=x.ForecastYear:
        m=x.Rate
        adjustedforecast=m*(x.Year_ofObservation-x.ForecastYear)+x.TrafficForecast
    else:
        adjustedforecast=x.TrafficForecast
    return adjustedforecast
```
