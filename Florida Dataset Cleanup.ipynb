```python
def find_PID(value): #to clean up the data
    if isinstance(value, str):
        end = value.find('--')
        if end == -1:
            return None
        else:
            start = value.find('-')+1
            return int(value[start:end])
```

```python
def TrafficForecast(x): #adjust the forecast
    if x.Year_ofObservation!=x.ForecastYear:
        m=x.Rate
        adjustedforecast=m*(x.Year_ofObservation-x.ForecastYear)+x.TrafficForecast
    else:
        adjustedforecast=x.TrafficForecast
    return adjustedforecast
```
