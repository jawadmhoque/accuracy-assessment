# Adjust Type 1
## Scale up the forecast to recent Year.
### use y=mx+c
$m=\frac{OYADTForecast-DYADTForecast}{OpeningYear-DesignYear}$

$adjusted count=m*(RecentYr-OpeningYear)+OYADTForecast$

```python
ohio_scaletype1=ohio_scaletype1.assign(m=(ohio_scaletype1['DYADTForecast']-ohio_scaletype1['OYADTForecast'])/
                                            (ohio_scaletype1['DesignYear']-ohio_scaletype1['OpeningYear']))
ohio_scaletype1[ohio_scaletype1['m']==0].head()
```

```python
ohio_scaletype1=ohio_scaletype1.assign(AdjustedForecast=ohio_scaletype1['m']*(ohio_scaletype1['RecentYr']-ohio_scaletype1['OpeningYear'])
                                      +ohio_scaletype1['OYADTForecast'])
ohio_scaletype1.head()
```
