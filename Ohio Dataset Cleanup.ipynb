## Explanation

If Planned Opening Year (2005) is greater than Actual Opening Year (2004), that means the project is planned for 2005 but opened in 2004. Forecast is available for 2005 but the quesiton is which actual count to take.

If the Near Actual Open Year (2003) is less than Actual Opening Year (2004), that means Near Actual Open Year Count is before the project has been opened and therefore cannot be compared against. In this case the actual year should be Recent Year. 
Scale Forecast to Recent Year.

If Near Actual Open Year (2005) is greater than Actual Opening Year (2004), that means the count can be used. But compare diffAOY_NAOY with diffRY_AOY. The smaller value should be the year.
